```python
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
from matplotlib import style
```

```python
randomNumbers = np.random.rand(10)
print(randomNumbers)
```

```python
style.use('ggplot')
plt.plot(randomNumbers,'g',color='b',label='line',linewidth=2, alpha=.4)
plt.xlabel('Range') # type: ignore
plt.ylabel('Numbers')
plt.title('First plot')
plt.annotate('Max',ha='center',va='bottom',xytext=(8,1500),xy=(11,1630),arrowprops = {'facecolor' : 'green'})
plt.legend()
plt.show()
```
